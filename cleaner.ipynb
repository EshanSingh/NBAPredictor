```python
# Imports
import pandas as pd
import numpy as np
```

```python
mvps = pd.read_csv('mvp.csv')
players = pd.read_csv('players.csv')
```

```python
# Select columns not present in players.csv
mvps = mvps[["Player","Year", "Pts Won", "Pts Max", "Share"]]
```

```python
players.drop(columns=['Rk'], inplace=True)
```

```python
# Cleaning names
players["Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
# Function to group players into same row
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Team"].str.endswith("TM")]
        row["Team"] = df.iloc[-1,:]["Team"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
complete = players.merge(mvps, on=["Player", "Year"], how="outer")
```

```python
complete.head()
```

```python
complete.drop(columns=['Rk'], inplace=True)
```

```python
# Fill NaN values with 0
complete[["Pts Won", "Pts Max", "Share"]] = complete[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
complete = complete[~complete["Player"].str.contains("League Average")]
```

```python
complete.head()
```

```python
teams = pd.read_csv('teams.csv')
```

```python
teams["Name"] = teams["Name"].str.replace("*", "", regex=False)
```

```python
# Clean Division rows
teams = teams[~teams["Name"].str.contains("Division")]
```

```python
teams["Team"] = teams["Name"]
```

```python
teams.drop(columns=["Name"], inplace=True)
teams.head()
```

```python
complete[complete["Player"] == "Nikola Jokić"]
```

```python
abb = {}

with open("abbreviations.csv") as file:
    lines = file.readlines()

    for line in lines[1:]:
        line = line.strip().replace("\n", "").split(",")
        abb[line[0]] = line[1]
```

```python
complete["Team"] = complete["Team"].map(abb)
complete.head()
```

```python
stats = complete.merge(teams, how="outer", on=["Team", "Year"])
stats.head()
```

```python
stats = stats.apply(pd.to_numeric, errors='ignore')
```

```python
stats['GB'] = stats['GB'].fillna(0)
stats['GB'] = stats['GB'].str.replace('—', '0.0')
stats['GB'].unique()
```

```python
stats['GB'] = pd.to_numeric(stats['GB'])
stats['GB'].dtype
```
